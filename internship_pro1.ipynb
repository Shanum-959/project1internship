# Step 1: Import Required Library
# Step 2: Load the Dataset

```python
import pandas as pd

df = pd.read_csv("data.csv")
```

# Step 3: Display the First Five Rows
# Step 4: Check Dataset Shape
# Step 5: Display Dataset Information
# Step 6: Display Statistical Summary

```python
print(df.head())

print(df.shape)

print(df.info())

print(df.describe())
```

# Step 7: Check Missing Values

```python
print(df.isnull().sum())
```

# Step 8: Calculate Missing Value Percentage

```python
missing_percentage = (df.isnull().sum()/len(df))*100

print(missing_percentage)
```

# Step 9: Handle Missing Values

```python
df["CouponCode"] = df["CouponCode"].fillna("No Coupon")
```

# Step 10: Verify Missing Values After Cleaning

```python
print(df.isnull().sum())
```

# Step 11: Check Duplicate Rows


```python
print("Duplicate Rows:", df.duplicated().sum())
```

# Step 12: Check Duplicate Order IDs

```python
print("Duplicate Order IDs:", df["OrderID"].duplicated().sum())
```

# Step 13: Convert Date Column into Datetime Format
# Step 14: Check Invalid Dates


```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

print("Invalid Dates:", df["Date"].isnull().sum())
```

# Step 15: Verify Data Types

```python
print(df.dtypes)
```

# Step 16: Verify Total Price

```python
df["CalculatedPrice"] = (df["Quantity"] * df["UnitPrice"]).round(2)

price_errors = df[df["CalculatedPrice"] != df["TotalPrice"]]

print(price_errors)
```

# Step 17: Remove Temporary Column

```python
df.drop("CalculatedPrice", axis=1, inplace=True)
```

# Step 18: Save the Cleaned Dataset

```python
df.to_csv("Cleaned_Orders_Dataset.csv", index=False)

print("Dataset cleaned and saved successfully!")
```
